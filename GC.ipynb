```python
import sys
```

```python
import gc
```

```python
# gc.disable() - GC 비활성화
# gc.enable() - GC 활성화
```

```python
# 순환참조하는 경우 GC가 발생되지 않음

obj1 = [1]
obj2 = [obj1]
obj1.append(obj2)

print(sys.getrefcount(obj1)) # obj2를 참조하기 때문에 3 출력
print(sys.getrefcount(obj2)) # obj1을 참조하기 때문에 3 출력
```

```python
# gc.collect() 를 사용해 명시적으로 메모리 해제 해제된 메모리 수 출력 

gc.collect() # 2 만큼의 메모리가 해제되어 2 출력
```

```python
gc.get_referrers(obj1) # obj1을 참조하는 모든 구조체, 객체 출력
```
